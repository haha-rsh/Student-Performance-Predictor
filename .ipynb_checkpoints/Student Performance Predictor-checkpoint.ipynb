## <center> EXPLORATORY DATA ANALYSIS

```python
import pyspark
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('BDA').getOrCreate()
```

```python
spark
```

```python
df = spark.read.csv('Dataset/Student_Performance.csv', header = True, inferSchema = True)
```

```python
df.show(5)
```

```python
type(df)
```

```python
df.printSchema()
```

```python
df.dtypes
```

```python
df.describe().show()
```

```python
df.columns
```

```python
cols = ['Hours Studied',
 'Previous Scores',
 'Extracurricular Activities',
 'Sleep Hours',
 'Sample Question Papers Practiced',
 'Performance Index']
```

```python
from pyspark.sql.functions import col,isnan,when,count
df_Columns=df.columns
df.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in df_Columns]).show()
```

```python
df \
    .groupby(cols) \
    .count() \
    .where('count > 1') \
    .sort('count', ascending=False) \
    .show()
```

```python
df.select(cols).distinct().count() 
```

```python
df.select(cols).count()
```

```python
df.groupBy('Extracurricular Activities').mean().show()
```

```python
df.groupBy('Hours Studied').mean().sort('Hours Studied', ascending=False).drop('avg(Hours Studied)').show()
```

```python
df.agg({'Performance Index':'avg'}).show()
```

```python
from pyspark.sql.functions import avg
numeric_columns = [col for col, dtype in df.dtypes if dtype in ['int', 'double', 'float', 'long']]
df.select([avg(col).alias(f"avg({col})") for col in numeric_columns]).show()
```

```python
df.filter(df['Sleep Hours']<=4).count()
```

```python
df.filter(df['Performance Index']>=99).show()
```

## Visualizing Data

```python
df_pandas = df.toPandas()
```

```python
df_pandas.columns
```

```python
df_pandas['Hours Studied'].describe()
```

```python
fig = px.histogram(df_pandas, x='Hours Studied', marginal='box', nbins=47, title='Distribution of Hours Studied')
fig.update_layout(bargap=0.1)
```

```python
fig = px.histogram(df_pandas, x='Previous Scores', marginal='box', color_discrete_sequence=['red'], title='Distribution of Previous Scores')
fig.update_layout(bargap=0.1)
```

```python
fig = px.histogram(df_pandas, x='Sleep Hours', marginal='box', color_discrete_sequence=['green'], title='Distribution of Sleep Hours')
fig.update_layout(bargap=0.1)
```

```python
fig = px.histogram(df_pandas, x='Performance Index',marginal='box', color='Extracurricular Activities', color_discrete_sequence=px.colors.qualitative.G10, title='Performance Index')
fig.update_layout(bargap=0.1)
```

```python
px.histogram(df_pandas, x='Extracurricular Activities', color='Sleep Hours',title='Extracurricular Activities',color_discrete_sequence=px.colors.qualitative.Pastel2)
```

```python
plt.scatter(df_pandas['Hours Studied'], df_pandas['Performance Index'])
plt.xlabel('Hours Studied')
plt.ylabel('Performance Index')
plt.title('Hours Studied vs. Performance Index')
plt.show()
```

```python
sns.boxplot(data=df_pandas,x='Previous Scores',hue='Extracurricular Activities',palette=['goldenrod','magenta'])
```

```python
plt.pie(df_pandas["Extracurricular Activities"].value_counts(), labels = df_pandas["Extracurricular Activities"].value_counts().index,
        shadow = True, autopct = "%1.1f%%")
plt.show()
```

```python
sns.boxplot(data=df_pandas, x='Hours Studied',y='Performance Index',hue='Hours Studied',legend=False, palette=['#636EFA', '#EF553B', '#00CC96', '#AB63FA', '#FFA15A', '#19D3F3', '#FF6692', '#B6E880', '#FF97FF'])
```

```python
avg_performance_by_hours = df_pandas.groupby('Hours Studied')['Performance Index'].mean()
plt.plot(avg_performance_by_hours.index, avg_performance_by_hours.values)
plt.show()
```
